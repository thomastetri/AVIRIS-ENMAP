# AVIRIS-NG L2 Surface Reflectance, Facility Instrument Collection, V1

```
https://www.earthdata.nasa.gov/data/catalog/ornl-cloud-aviris-ng-l2-reflectance-2110-1
https://cmr.earthdata.nasa.gov/virtual-directory/collections/C2659129205-ORNL_CLOUD
```

````python
import requests

url = "https://cmr.earthdata.nasa.gov/search/granules.json"
params = {
    "collection_concept_id": "C2659129205-ORNL_CLOUD",
    "page_size": 1000,
    "page_num": 1,
    "sort_key": "-start_date",
    "temporal": "2022-07-01T00:00:00.000Z,"
}

response = requests.get(url, params=params)

# Raise an error if something went wrong
response.raise_for_status()

# Parse JSON into a Python dictionary
data = response.json()

# Example: print how many granules you got
granules = len(data.get('feed', {}).get('entry', []))
print(f"Got {granules} granules, aka {granules // 2} images (.bin + .hdr files)") 

# Access the whole dictionary if you want
granules = data['feed']['entry']
granules = sorted(granules, key=lambda x: x['title'])
````

After downloading consider converting to GeoTiff or CoG, e.g.:

```
gdalwarp -f GTiff -co bigtiff=yes -co compress=lzw ang20220917t211723_rfl_v2aa1a_img.bin ang20220917t211723_rfl_v2aa1a_img.tif
```
